```python
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np

fig,ax = plt.subplots(2,3,figsize=(16,8))

n = 10
p= [[0.1,0.3,0.4],
    [0.5,0.6,0.9]]
r_vals = list(range(11))
for i in range(2):
    for j in range(3):
        #print(i,j,p[i][j])
        prob_vals = [binom.pmf(r,n,p[i][j]) for r in r_vals]

        #print(prob_vals)
        r_new = np.linspace(min(r_vals),max(r_vals),200)
        spl = make_interp_spline(r_vals,prob_vals,k=3)
        y_smooth = spl(r_new)

        
        ax[i,j].plot(r_new,y_smooth)
        ax[i,j].set_title("n = %s  p = %s" % (n,p[i][j]))

fig.suptitle("Binomial distribution")
plt.tight_layout()
plt.show()
        
```
