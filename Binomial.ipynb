### Binomial Distribution
The binomial distribution qith paramters $n$ and $p$ is a discrete probability distribution of number of successes in a sequence of n IID Bernoulli trials, each asking a yes-no question (binary outcome).

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

fig,ax = plt.subplots(2,3,figsize=(16,8))

n = 10
p= [[0.1,0.3,0.4],
    [0.5,0.6,0.9]]
r_vals = list(range(11))
for i in range(2):
    for j in range(3):
        #print(i,j,p[i][j])
        prob_vals = [binom.pmf(r,n,p[i][j]) for r in r_vals]

        #print(prob_vals)
        r_new = np.linspace(min(r_vals),max(r_vals),200)
        spl = make_interp_spline(r_vals,prob_vals,k=3)
        y_smooth = spl(r_new)

        
        ax[i,j].plot(r_new,y_smooth)
        ax[i,j].set_title("n = %s  p = %s" % (n,p[i][j]))

fig.suptitle("Binomial distribution")
plt.tight_layout()
plt.show()
        
```

```python
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

fig,ax = plt.subplots(1,1)

n = 10
p= [[0.1,0.3,0.4],
    [0.5,0.6,0.9]]
r_vals = list(range(11))
for i in range(2):
    for j in range(3):
        alp = 0.8
        prob_vals = [binom.pmf(r,n,p[i][j]) for r in r_vals]

        #print(prob_vals)
        r_new = np.linspace(min(r_vals),max(r_vals),200)
        spl = make_interp_spline(r_vals,prob_vals,k=3)
        y_smooth = spl(r_new)

        
        ax.plot(r_new,y_smooth,alpha=alp,label=f"n = {n}  p= {p[i][j]}")
        ax.set_xlabel("X")
        ax.set_ylabel("P(X)")
        ax.legend()
        ax.set_xticks(r_vals)
        alp -= 0.15

fig.suptitle("Binomial distribution")
plt.tight_layout()
plt.show()
        
```
