### Geometric Distribution
The probability distribution of $X$ number of bernoulli trials needed before getting first success, supported on the set $\{0,1,2,...\}$




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline,BSpline

fig,ax = plt.subplots(2,2,figsize=(10,9))
x = list(range(6))
p = [0.2,0.5,0.6,0.8]
def geom(x,p):
    f = -1
    g = -1
    for i in range(2):
        for j in range(2):
            g += 1
            f += 1
            prob_vals = [[round(pow((1-k),l)*k,2) for l in x] for k in p]

            x_new  = np.linspace(min(x),max(x),200)
            spl = make_interp_spline(x,prob_vals[f],k=3)
            y_smooth = spl(x_new)

            ax[i,j].plot(x_new,y_smooth,color="magenta")
            ax[i,j].set_title("p = %.2f" % (p[g]))
        
geom(x,p)    
fig.suptitle("Geometric Distribution")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline,BSpline

fig,ax = plt.subplots(1,1)
x = list(range(6))
p = [0.2,0.5,0.6,0.8]
def geom(x,p):
    f = -1
    g = -1
    alp = 0.8
    for i in range(2):
        for j in range(2):
            g += 1
            f += 1
            prob_vals = [[round(pow((1-k),l)*k,2) for l in x] for k in p]

            x_new  = np.linspace(min(x),max(x),200)
            spl = make_interp_spline(x,prob_vals[f],k=3)
            y_smooth = spl(x_new)

            ax.plot(x_new,y_smooth,label=f"p = {p[g]}",alpha=alp)
            ax.set_xlabel("X")
            ax.set_ylabel("P(X)")
            ax.legend()
            alp -= 0.15
        
geom(x,p)    
fig.suptitle("Geometric Distribution")
```
