### Discrete Uniform Distribution
the discrete uniform distribution is a symmetric probability distribution wherein a finite number of values are equally likely to be observed; every one of the $n$ values has equal probability $1/n$ <br>
Another way of saying "discrete uniform distribution" would be "a known, finite number of outcomes equally likely to happen"

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline,BSpline

fig,ax = plt.subplots(1,1)
a = 2
b = 7

def dunif(a,b):
    n = b-a+1
    x = list(range(a,b+1))
    prob_vals = [1/n for k in x]

    ax.stem(x,prob_vals)
    ax.set_title("Discrete Uniform Distribution")
    ax.set_xlabel("X")
    ax.set_ylabel("P(X)")
    ax.set_xticks(x)
    ax.set_ylim(0,0.3)
    ax.set_xlim(0,10)
dunif(a,b)
```
