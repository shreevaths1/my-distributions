### Negative Binomial Distribution
The negative binomial distribution is a discrete probability distribution that models the number of failures in a sequence of IID Bernoulli trials before a specified number of successes (denoted $r$) occurs

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline,BSpline

fig,ax = plt.subplots(2,3,figsize=(15,10))

r = list(range(1,7))
x = list(range(1,11))
p = 0.5

def nbinom(r,x,p):
    f = -1
    for i in range(2):
        for j in range(3):
            f += 1
            prob_vals = [round((math.comb(a-1,r[f]-1))*(pow(p,r[f]))*(pow(1-p,a-r[f])),2) for a in list(range(r[f],r[f]+10))]

            x_new = np.linspace(min(x),max(x),200)
            spl = make_interp_spline(x,prob_vals,k=3)
            y_smooth = spl(x_new)

            ax[i,j].plot(x_new,y_smooth)
            ax[i,j].set_title("r = %d" % r[f])

fig.suptitle("Negative Binomial")
nbinom(r,x,p)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline,BSpline

fig,ax = plt.subplots(1,1)

r = list(range(1,7))
x = list(range(1,11))
p = 0.5

def nbinom(r,x,p):
    f = -1
    alp = 0.8
    for i in range(2):
        for j in range(3):
            f += 1
            prob_vals = [round((math.comb(a-1,r[f]-1))*(pow(p,r[f]))*(pow(1-p,a-r[f])),2) for a in list(range(r[f],r[f]+10))]

            x_new = np.linspace(min(x),max(x),200)
            spl = make_interp_spline(x,prob_vals,k=3)
            y_smooth = spl(x_new)

            ax.plot(x_new,y_smooth,alpha=alp,label=f"r = {r[f]}")
            ax.set_xlabel("X")
            ax.set_ylabel("P(X)")
            ax.legend()
            alp -= 0.07
            

fig.suptitle("Negative Binomial")
nbinom(r,x,p)
```
