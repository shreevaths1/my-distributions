### Poisson Distribution
Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event.




```python
import math
import matplotlib.pyplot as plt
from itertools import repeat
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np

fig,ax = plt.subplots(2,2,figsize=(10,9))


x= [0,1,2,3,4,5]
def poiss(x):
    mu = 1
    for i in range(2):
        for j in range(2):
            
            a=[pow(mu,k) for k in x]
            b = list(repeat(math.exp(-mu),6))
            c=[math.factorial(k) for k in x]
            prob_vals = [((p*q)/r) for p,q,r in zip(a,b,c)]

            x_new = np.linspace(min(x),max(x),200)
            spl = make_interp_spline(x,prob_vals,k=3)
            y_smooth = spl(x_new)
            
            ax[i,j].plot(x_new,y_smooth,color="green")
            ax[i,j].set_title("mu = %s"% mu)
            mu+= 1

fig.suptitle("Poisson Distribution")
poiss(x)
# ax[0,0].plot()
```

```python
import math
import matplotlib.pyplot as plt
from itertools import repeat
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np

fig,ax = plt.subplots(1,1)


x= [0,1,2,3,4,5]
def poiss(x):
    mu = 1
    alp = 0.8
    for i in range(2):
        for j in range(2):
            
            a=[pow(mu,k) for k in x]
            b = list(repeat(math.exp(-mu),6))
            c=[math.factorial(k) for k in x]
            prob_vals = [((p*q)/r) for p,q,r in zip(a,b,c)]

            x_new = np.linspace(min(x),max(x),200)
            spl = make_interp_spline(x,prob_vals,k=3)
            y_smooth = spl(x_new)
            
            ax.plot(x_new,y_smooth,label=f"mu = {mu}",alpha=alp)
            ax.set_xlabel("X")
            ax.set_ylabel("P(X)")
            ax.legend()
            alp -= 0.15
            mu+= 1

fig.suptitle("Poisson Distribution")
poiss(x)
# ax[0,0].plot()
```
