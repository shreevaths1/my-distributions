### Hypergeomteric Distribution
The hypergeometric distribution is a discrete probability distribution that describes the probability of $k$ (random draws for which the object drawn has a specified feature) in $n$ draws, without replacement, from a finite population of size $N$  that contains exactly $K$ objects with that feature, wherein each draw is either a success or a failure.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline,BSpline
from math import comb

fig,ax = plt.subplots(1,3,figsize=(15,5))
N = [500,500,500]
K = [50,60,70]
n = [100,200,300]
x = list(range(60))
def hypergeom(N,n,K,x):
    f = -1
    for i in range(1):
        for j in range(3):
            f += 1
            prob_vals = [round((comb(K[f],i)*comb(N[f]-K[f],n[f]-i))/(comb(N[f],n[f])),2) for i in x]

            x_new = np.linspace(min(x),max(x),200)
            spl = make_interp_spline(x,prob_vals,k=3)
            y_smooth = spl(x_new)

            ax[j].plot(x_new,y_smooth,linestyle="dashdot",dash_capstyle="round",linewidth=4)
            ax[j].set_ylim(0,0.17)
            ax[j].set_yticks(list(np.arange(0,0.17,0.01)))
            
fig.suptitle("Hypergeometric Distribution")
hypergeom(N,n,K,x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline,BSpline
from math import comb

fig,ax = plt.subplots(1,1,figsize=(15,5))
N = [500,500,500]
K = [50,60,70]
n = [100,200,300]
x = list(range(60))
def hypergeom(N,n,K,x):
    f = -1
    for i in range(1):
        for j in range(3):
            f += 1
            alp = 0.8
            prob_vals = [round((comb(K[f],i)*comb(N[f]-K[f],n[f]-i))/(comb(N[f],n[f])),2) for i in x]

            x_new = np.linspace(min(x),max(x),200)
            spl = make_interp_spline(x,prob_vals,k=3)
            y_smooth = spl(x_new)

            ax.plot(x_new,y_smooth,linestyle="dashdot",dash_capstyle="round",linewidth=2.5,alpha= alp,label= f"N = {N[f]} K = {K[f]} n = {n[f]}")
            ax.set_ylim(0,0.17)
            ax.set_yticks(list(np.arange(0,0.17,0.01)))
            ax.set_xlabel("X")
            ax.set_ylabel("P(X)")
            ax.legend()
            alp -= 0.30
            
fig.suptitle("Hypergeometric Distribution")
hypergeom(N,n,K,x)
```
